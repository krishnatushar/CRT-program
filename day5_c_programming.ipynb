```
//read a limit of numbers and print the count of even palindromes in that limit 
#include<stdio.h>
int isPalindrome(int n)
{
	int rev=0,buffer=n;
	while(n!=0){
		rev=rev*10+(n%10);
		n/=10;
	}
	if(buffer==rev){
		return 1;
	}
	else{
		return 0;
	}
}
int main()
{
int i,a1,a2,res,pcnt=0;
scanf("%d%d",&a1,&a2);
for(i=a1;i<=a2;i++)
{
	if(i%2==0)
	{
		res=isPalindrome(i);
		if(res==1){
			pcnt++;
		}
	}
}
printf("%d",pcnt);
return 0;
}
```

```
//read the input and display factorial digit sum
#include<stdio.h>
int factorial(int x)
{
	int i,fact=1;
	for(i=x;i>0;i--)
	{
		fact *=i;
	}
	return fact;
}
int main()
{
	int n,r,s=0,buffer;
	printf("enter the number");
	scanf("%d",&n);
	buffer=n;
	while(n!=0)
	{
		r=n%10;
		s=s+factorial(r);
		n/=10;
	}
	if(s==buffer)
	{
		printf("yes");
	}
	else
	{
		printf("no");
	}
	return 0;
}
```

```
/read the input as two limits and print the count of perfect numbers
#include<stdio.h>
int isPerfectNumber(int n)
{
	int i,sum=0;
	for(i=1;i<=(n/2);i++)
	{
		if(n%i==0)
		{
			sum=sum+i;
		}
    }
    if(sum==n)
		{
			printf("%d ",sum);
			return 1;
		}
		else
		{
			return 0;
		}
}
int main()
{
	int a1,a2,i,res,perfectCnt=0;
	printf("enter the numbers");
	scanf("%d%d",&a1,&a2);
	for(i=a1;i<=a2;i++)
	{
		res=isPerfectNumber(i);
		if(res==1)
		{
			perfectCnt++;
		}
	}
	printf("%d",perfectCnt);
	return 0;
}
```

```
//read an input and output as sum of each digits to power of digit count
#include<stdio.h>
#include<math.h>
int digitCount(int n)
{
	int cnt=0;
	while (n!=0)
	{
		cnt++;
		n/=10;
	}
	return cnt;
}
int main()
{
	int x,dgtcnt,r,sum=0;
	scanf("%d",&x);
	dgtcnt=digitCount(x);
	while(x!=0)
	{
		r=x%10;
		x=x/10;
		sum=sum+pow(r,dgtcnt);
	}
	printf("%d",sum);
	return 0;
}
```

```
//read the limit and print the prime numbers
#include<stdio.h>
int isPrime(n)
{
	int i,flag=0;
	for(i=2;i<=(n/2);i++)
	{
	if(n%i==0)
	  {
		flag=1;
		break;
	  }
	}
	if(flag==0)
	{
		return 1;
	}
	else
	{
		return 0;
	}
}
int main()
{
	int a1,a2,i,res;
	scanf("%d%d",&a1,&a2);
	for(i=a1;i<=a2;i++)
	{
		res=isPrime(i);
		if(res==1)
		{
			printf("%d ",i);
		}
	}
	return 0;
}
```
